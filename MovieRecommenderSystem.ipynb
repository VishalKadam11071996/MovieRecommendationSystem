```python
# Extract name from object (genres)
def convert(obj):
  """
  Input: obj - This will be string type ex: '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, 
                                            {"id": 878, "name": "Science Fiction"}]'
  Output: list of name
  """
  op = []
  for d in ast.literal_eval(obj): # ast.literal_eval will covert string to list object
    op.append(d['name'])
  return op
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Extract name from object (cast)
def extractCast(obj):
  """
  Input: obj - This will be string type ex: '[{"cast_id": 242, "character": "Jake Sully", "credit_id": "5602a8a7c3a3685532001c9a", "gender": 2,
                                            "id": 65731, "name": "Sam Worthington", "order": 0}, {"cast_id": 3, "character": "Neytiri", 
                                            "credit_id": "52fe48009251416c750ac9cb", "gender": 1, "id": 8691, "name": "Zoe Saldana", "order": 1},
                                            {"cast_id": 25, "character": "Dr. Grace Augustine", "credit_id": "52fe48009251416c750aca39", "gender": 1, 
                                            "id": 10205, "name": "Sigourney Weaver", "order": 2}'
  Output: list of top 3 Casting name
  """
  op = []
  count = 0
  for d in ast.literal_eval(obj): # ast.literal_eval will covert string to list object
    if count == 3:
      break
    op.append(d['name'])
    count += 1
  return op
```

```python
movies['cast'] = movies['cast'].apply(extractCast)
movies.head()
```

```python
movies.iloc[0].crew
```

```python
# Extract director name from object
def extractDirector(obj):
  """
  Input: obj - This will be string type
  Output: list of top 3 Casting name
  """
  op = []
  for d in ast.literal_eval(obj): # ast.literal_eval will covert string to list object
    if d['job'] == "Director":
      op.append(d['name'])
      break
  return op
```

```python
movies['crew'] = movies['crew'].apply(extractDirector)
movies.head()
```

```python
# lets convert overview column from string to list so we can append that in combined feature.
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
# Remove sapce between columns
# Reason - Vishal Kadam or Vishal Joshi if we keep space then module mi8 confuse by Vishal so
# if we remove space then it willl be VishalKadam and VishalJoshi, using this model will not confuse 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
# Cobined Features in one column
movies['combined-feature'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies_df = movies[['movie_id', 'title', 'combined-feature']]
new_movies_df['combined-feature'][0]
```

```python
# converting combined-feaature from list to single string.
new_movies_df['combined-feature'] = new_movies_df['combined-feature'].apply(lambda x: " ".join(x))
```

```python
# converting string to lower case
new_movies_df['combined-feature'] = new_movies_df['combined-feature'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english') #here max features means 5000 words columns
```

Calculate Cosign Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity # 0 is low and 1 is high similarity
```

```python
similarity_mtx = cosine_similarity(vectors) # 4806 movies are present so matrix will be 4806 * 4806
```

```python
# below is vector of 1st movie with smilarity of other 4806 movies,
# here 1 indicate the similarity with self movie
similarity_mtx[0]
```

Recommend Movie

```python
def movieRecommender(movie_name):
  movie_index = new_movies_df[new_movies_df['title'] == movie_name].index[0]
  distances = similarity_mtx[movie_index]
  top_5_similar_movie = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

  for movie in top_5_similar_movie:
    print(new_movies_df.iloc[movie[0]].title)
```

```python
movieRecommender('Avatar')
```
